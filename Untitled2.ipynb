```python
import random

## Character í´ë˜ìŠ¤
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.hp = hp
        self.level = level
        self.attack = attack
        self.defense = defense

    # ì¸ìŠ¤í„´ìŠ¤ì˜ í˜„ì¬ ì²´ë ¥ì´ 0 ì´ìƒì¸ì§€ë¥¼ í™•ì¸
    def is_alive(self):
        return self.hp > 0

    # ê³µê²© ë°›ì•˜ì„ ë•Œ ì²´ë ¥ê°ì†Œ
    def take_damage(self, damage):
        real_damage = max(damage - self.defense, 0)
        self.hp -= real_damage
        print(f"ğŸ§ğŸ»ğŸ§ğŸ»â€â™€ï¸ {self.name}(ì´)ê°€ {real_damage} ë§Œí¼ í”¼í•´ë¥¼ ì…ì—ˆìŠµë‹ˆë‹¤!\n >>> ë‚¨ì€ ì²´ë ¥: {self.hp}")

    # ë‚´ê°€ ê³µê²©í•˜ê¸°
    def attack_target(self, target):
        my_attack = random.randint(1, self.attack)
        print(f"ğŸ§Ÿâ€â™€ï¸ğŸ§Ÿâ€â™€ï¸ğŸ§Ÿâ€â™€ï¸ {self.name}(ì´)ê°€ {target.name}ì—ê²Œ {my_attack}ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤!")
        target.take_damage(my_attack)


## í”Œë ˆì´ì–´ í´ë˜ìŠ¤
class Player(Character):
    def __init__(self, name):
        super().__init__(name, 1, 100, 25, 5)  # âœ… ì´ˆê¸°ê°’ ì„¤ì •
        self.experience = 0  # âœ… ê²½í—˜ì¹˜ê°€ 0ìœ¼ë¡œ ì‹œì‘

    # ê²½í—˜ì¹˜ íšë“
    def gain_experience(self, amount):
        self.experience += amount
        print(f'ğŸ§šğŸ»â€â™‚ï¸ {self.name}ê°€ ê²½í—˜ì¹˜ {amount}ë¥¼ íšë“í•˜ì˜€ìŠµë‹ˆë‹¤.\n >>> í˜„ì¬ ê²½í—˜ì¹˜ : {self.experience}')

    # ë ˆë²¨ ì—…
    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience = 0  # âœ… ê²½í—˜ì¹˜ ì´ˆê¸°í™”
        print(f'ğŸ§šğŸ»â€â™‚ï¸ ë ˆë²¨ì—…! ğŸ‰ğŸ‰ \n í˜„ì¬ ë ˆë²¨: {self.level}\n ê³µê²©ë ¥: {self.attack}\n ë°©ì–´ë ¥: {self.defense}')


## ëª¬ìŠ¤í„° í´ë˜ìŠ¤
class Monster(Character):
    def __init__(self, name, level):
        super().__init__(name, level,
                         random.randint(10, 30) * level,
                         random.randint(5, 20) * level,
                         random.randint(1, 5) * level)


## battle ì‹œì‘
def Battle(player, monster):
    print(f"ğŸ§ğŸ» {player.name} vs ğŸ§›â€â™‚ï¸ {monster.name} ì „íˆ¬ ì‹œì‘!")

    # ì „íˆ¬ ì‹œì‘ ì „, ìƒíƒœ ì¶œë ¥
    print(f"ì´ë¦„ : {player.name}     ì´ë¦„ : {monster.name}")
    print(f"ë ˆë²¨ : {player.level}     ë ˆë²¨ : {monster.level}")
    print(f"ì²´ë ¥ : {player.hp}       ì²´ë ¥ : {monster.hp}")
    print(f"ê³µê²©ë ¥ : {player.attack}  ê³µê²©ë ¥ : {monster.attack}")
    print(f"ë°©ì–´ë ¥ : {player.defense} ë°©ì–´ë ¥ : {monster.defense}\n")

    while player.is_alive() and monster.is_alive():
        # í”Œë ˆì´ì–´ ê³µê²©
        print(f"\nâš” {player.name}ì˜ í„´ì…ë‹ˆë‹¤.")
        damage = random.randint(1, player.attack)
        print(f"{monster.name}(ì´)ê°€ {damage}ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…ì—ˆìŠµë‹ˆë‹¤.")
        before_hp = monster.hp
        monster.take_damage(damage)
        print(f"{before_hp} -> {monster.hp}")  # ì²´ë ¥ ë³€í™” í‘œì‹œ
        print(f"{monster.name}ì˜ ë‚¨ì€ ì²´ë ¥ : {monster.hp}\n")

        if not monster.is_alive():
            print(f"\nğŸ‰ {player.name} ìŠ¹ë¦¬!")
            exp_gain = monster.level * 20
            print(f"{exp_gain}ë§Œí¼ì˜ ê²½í—˜ì¹˜ë¥¼ íšë“í–ˆìŠµë‹ˆë‹¤.")
            player.gain_experience(exp_gain)
            return

        # ëª¬ìŠ¤í„° ê³µê²© (âœ… ëª¬ìŠ¤í„°ê°€ í”Œë ˆì´ì–´ë¥¼ ê³µê²©í•˜ë„ë¡ ìˆ˜ì •!)
        print(f"\nâš” {monster.name}ì˜ í„´ì…ë‹ˆë‹¤.")
        damage = random.randint(1, monster.attack)
        print(f"{player.name}(ì´)ê°€ {damage}ë§Œí¼ì˜ í”¼í•´ë¥¼ ì…ì—ˆìŠµë‹ˆë‹¤.")
        before_hp = player.hp
        player.take_damage(damage)  # âœ… ì˜¬ë°”ë¥´ê²Œ `player`ë¥¼ ê³µê²©í•˜ë„ë¡ ìˆ˜ì •!
        print(f"{before_hp} -> {player.hp}")  # ì²´ë ¥ ë³€í™” í‘œì‹œ
        print(f"{player.name}ì˜ ë‚¨ì€ ì²´ë ¥ : {player.hp}\n")

        if not player.is_alive():
            print(f"\nğŸ’€ {player.name} íŒ¨ë°°...")
            return


## Main í•¨ìˆ˜
def Main():
    # ëª¬ìŠ¤í„° ë¦¬ìŠ¤íŠ¸
    Monster_dict = {'ğŸ§Ÿâ€â™€ï¸ ì¢€ë¹„': 1, 'ğŸ§Œ íŠ¸ë¡¤': 2, 'ğŸ§›â€â™‚ï¸ ë±€íŒŒì´ì–´': 3}

    # í”Œë ˆì´ì–´ ìƒì„±
    player_name = input("ğŸ§šğŸ»â€â™‚ï¸ í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(player_name)
    print(f"\nğŸ§ğŸ» {player.name}ê°€ ëª¨í—˜ì„ ì‹œì‘í•©ë‹ˆë‹¤!\n")

    # ëª¬ìŠ¤í„°ì™€ ì‹¸ìš°ê¸°
    for monster_name, level in Monster_dict.items():
        monster = Monster(monster_name, level)
        print(f"\nâš” {monster.name} (ë ˆë²¨ {monster.level}) ì¶œí˜„! ì „íˆ¬ ì‹œì‘!")

        Battle(player, monster)

        # í”Œë ˆì´ì–´ê°€ ì£½ìœ¼ë©´ ì¦‰ì‹œ ê²Œì„ ì¢…ë£Œ
        if not player.is_alive():
            print("ğŸª¦ í”Œë ˆì´ì–´ê°€ ì‚¬ë§í•˜ì˜€ìŠµë‹ˆë‹¤ ğŸª¦")
            return

    # ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í•œ ê²½ìš°
    print(f"\nğŸ‰ ğŸ§šğŸ»â€â™‚ï¸ {player.name}, ëª¨ë“  ì „íˆ¬ì—ì„œ ìŠ¹ë¦¬í–ˆìŠµë‹ˆë‹¤! ëª¨í—˜ ì™„ë£Œ! ğŸ‰")

# âœ… `Main()` í•¨ìˆ˜ ì‹¤í–‰
Main()











```
